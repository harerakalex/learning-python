# Tesseract OCR

```python
# Function to read read text

import pytesseract

def read_text(image):
    pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
    text = pytesseract.image_to_string(image)
    
    return text
```

```python
team_img = Image.open('images/team.jpg')

plt.imshow(team_img)
```

```python
# search text on the image
# As you can see the text are not fully found
text = read_text(team_img)
print(text)
```

```python
# Try to transform image to see if we can get more detail about our image
croped_img = team_img.crop((140, 520, 1150, 720))
print(croped_img.size)
plt.imshow(croped_img)
# crop.size
```

```python
# Since I have found my text region, Now I can go ahead and search for my text
text = read_text(croped_img)
print(text)
```
