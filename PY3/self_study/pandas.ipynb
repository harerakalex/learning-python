```python
# Try to the above using group by
country_grp['HaveWorkedLanguage'].str.contains('Python').sum()
```

```python
# then I should use apply just get rid of the above error

country_grp['HaveWorkedLanguage'].apply(lambda x: x.str.contains('Python').sum())
```

```python
# Find percentage of people that use python in each country

country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['HaveWorkedLanguage'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_pd = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_pd
```

```python
python_pd.rename(columns={'Country': 'NumRespondents', 'HaveWorkedLanguage': 'NumKnowsPython'}, inplace=True)
python_pd
```

```python
python_pd['PctKnowsPyhton'] = (python_pd['NumKnowsPython']/python_pd['NumRespondents']) * 100
python_pd
```

```python
python_pd.sort_values(by='PctKnowsPyhton', ascending=False)
```
