# Manipulating DataFrames

```python
df = pd.read_csv('readonly/house_prices.csv')

df.head()
```

```python
df['first'] = df['street']
#using regex
df['first'] = df['first'].replace("[ ].*", "", regex=True)
df.head()
```

```python
# verify the above regex
import re
s = "hello world"

print(re.findall("[ ].*", s))
```

```python
del(df['first'])
df.head()
```

```python
def splitname(row):
    row['first'] = row['street'].split(" ")[0]
    row['last'] = row['street'].split(" ")[-1]
    return row

df = df.apply(splitname, axis = 'columns')
df.head()
```

```python
del(df['first'])
del(df['last'])
df.head()
```

```python
pattern = "(^[\w]*)(?:.* )([\w]*$)"
df['street'].str.extract(pattern).head()
```

```python
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"
names = df['street'].str.extract(pattern).head()
names
```

```python
df['First'] = names['First']
df['Last'] = names['Last']
df.head()
```

```python
pattern = "([\w]{3} [\w]{1,2}, [\w]{4})"
df['Born'] = df['street'].str.extract(pattern)
df['Born'].head()
```
